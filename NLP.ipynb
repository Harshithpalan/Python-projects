## Load Pre-trained Model

### Subtask:
Load a pre-trained generative Transformer model and its corresponding tokenizer from the `transformers` library.


### Loading the GPT-2 Model and Tokenizer

**Reasoning**:
In this step, we will load the pre-trained weights and the specific tokenizer required for the 'gpt2' model. The tokenizer converts text into numerical input that the model can understand, and the `GPT2LMHeadModel` is designed specifically for language modeling tasks like text generation.

**Reasoning**:
I will load the GPT-2 tokenizer and model using the transformers library as instructed to prepare for text generation.



```python
from transformers import GPT2LMHeadModel, GPT2Tokenizer

model_name = "gpt2"

# Load pre-trained tokenizer
tokenizer = GPT2Tokenizer.from_pretrained(model_name)

# Load pre-trained model
model = GPT2LMHeadModel.from_pretrained(model_name)

print(f"Successfully loaded tokenizer and model for: {model_name}")
```

## Generate Text

### Subtask:
Write a Python script to encode an input prompt, generate text using the GPT-2 model, and decode the output.


### Text Generation Process

**Reasoning**:
To generate text, we first need to convert a human-readable prompt into numerical tokens that the model can process. We then use the model's `generate` method to predict the subsequent tokens in the sequence. Finally, we convert these numerical tokens back into a human-readable string while removing special formatting tokens.

**Reasoning**:
I will now implement the Python script to encode the input prompt, generate a text sequence using the loaded GPT-2 model, and decode the output as per the instructions.



```python
# 1. Define input prompt
prompt = 'In a galaxy far, far away'

# 2. Encode input prompt to PyTorch tensors
input_ids = tokenizer.encode(prompt, return_tensors='pt')

# 3. Generate output sequences
output_sequences = model.generate(
    input_ids,
    max_length=50,
    num_return_sequences=1,
    pad_token_id=tokenizer.eos_token_id
)

# 4. Decode the generated output IDs
generated_text = tokenizer.decode(output_sequences[0], skip_special_tokens=True)

# 5. Print the generated text
print('--- Generated Output ---')
print(generated_text)
```
